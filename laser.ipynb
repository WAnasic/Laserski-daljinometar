```python
import numpy as np
import cv2 as cv
import glob

# Kriterijumi zaustavljanja iteracija
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)

# Tačke objekta (šahovske table) u realnom svetu (koordinate X i Y)
objp = np.zeros((6*7,3), np.float32)
objp[:,:2] = np.mgrid[0:7,0:6].T.reshape(-1,2)

# Liste za čuvanje tačaka objekta i tačaka slike
objpoints = []  # 3D tačke u realnom prostoru
imgpoints = []  # 2D tačke na ravni slike

# Pronalaženje svih JPEG slika u trenutnom direktorijumu
images = glob.glob('slika1.jpg')

for fname in images:
  # Učitavanje i konverzija slike u grayscale
  img = cv.imread(fname)
  gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

  # Pronalaženje uglova šahovske table
  ret, corners = cv.findChessboardCorners(gray, (7,6), None)

  # Dodavanje tačaka objekta i slike ako su uglovi pronađeni
  if ret == True:
    objpoints.append(objp.copy())
    corners2 = cv.cornerSubPix(gray,corners, (11,11), (-1,-1), criteria)
    imgpoints.append(corners2)

  # Crtež i prikaz uglova na slici
  cv.drawChessboardCorners(img, (7,6), corners2, ret)
  cv.imshow('img', img)
  cv.waitKey(500)

# Zatvaranje svih prozora nakon petlje (ne unutar if bloka)
cv.destroyAllWindows()
```
